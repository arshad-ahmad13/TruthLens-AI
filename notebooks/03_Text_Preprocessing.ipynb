```python
import pandas as pd
import numpy as np

import re
import string

import nltk

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer

import warnings
warnings.filterwarnings("ignore")
```

```python
nltk.download("punkt")
nltk.download("stopwords")
nltk.download("wordnet")
nltk.download("omw-1.4")
```

```python
nltk.download("punkt_tab")
```

```python
df = pd.read_csv("../data/processed/eda_completed.csv")

df.head()
```

```python
df = df[["title","text","label"]]

df.head()
```

```python
df["content"] = df["title"] + " " + df["text"]

df.head()
```

```python
df["content"] = df["content"].str.lower()

df.head()
```

```python
def remove_html(text):
    clean = re.compile("<.*?>")
    return re.sub(clean, "", text)
```

```python
df["content"] = df["content"].apply(remove_html)
```

```python
def remove_url(text):
    return re.sub(r"https?://\S+|www\.\S+", "", text)
```

```python
df["content"] = df["content"].apply(remove_url)
```

```python
def remove_numbers(text):
    return re.sub(r"\d+", "", text)
```

```python
df["content"] = df["content"].apply(remove_numbers)
```

```python
def remove_punctuation(text):
    return text.translate(
        str.maketrans("", "", string.punctuation)
    )
```

```python
df["content"] = df["content"].apply(remove_punctuation)
```

```python
def remove_spaces(text):
    return " ".join(text.split())
```

```python
df["content"] = df["content"].apply(remove_spaces)
```

```python
def remove_special(text):
    return re.sub(r"[^a-zA-Z\s]", "", text)
```

```python
df["content"] = df["content"].apply(remove_special)
```

```python
df["tokens"] = df["content"].apply(word_tokenize)

df.head()
```

```python
stop_words = set(stopwords.words("english"))
```

```python
def remove_stopwords(tokens):
    return [
        word
        for word in tokens
        if word not in stop_words
    ]
```

```python
df["tokens"] = df["tokens"].apply(remove_stopwords)
```

```python
stemmer = PorterStemmer()
```

```python
def stemming(tokens):
    return [
        stemmer.stem(word)
        for word in tokens
    ]
```

```python
df["stemmed"] = df["tokens"].apply(stemming)
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
def lemmatization(tokens):
    return [
        lemmatizer.lemmatize(word)
        for word in tokens
    ]
```

```python
df["lemmatized"] = df["tokens"].apply(lemmatization)
```

```python
df["clean_text"] = df["lemmatized"].apply(
    lambda x: " ".join(x)
)

df.head()
```

```python
print("Original Text\n")
print(df.loc[0,"text"])

print("\n"+"="*100+"\n")

print("Clean Text\n")
print(df.loc[0,"clean_text"])
```
